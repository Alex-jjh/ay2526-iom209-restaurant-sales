## C. Data Cleaning Summary

This section documents the missing values and inconsistencies found in the original dataset, and explains how they were handled.

```python
# Load the original (uncleaned) dataset
df_original = pd.read_csv('restaurant_sales_data.csv')

print("=" * 60)
print("ORIGINAL DATASET OVERVIEW")
print("=" * 60)
print(f"\nDataset Shape: {df_original.shape[0]} rows × {df_original.shape[1]} columns")
print(f"\nColumns: {list(df_original.columns)}")
print(f"\nData Types:")
print(df_original.dtypes)
```

```python
# Analyze missing values in the original dataset
print("\n" + "=" * 60)
print("MISSING VALUE ANALYSIS")
print("=" * 60)

missing_analysis = pd.DataFrame({
    'Column': df_original.columns,
    'Missing Count': df_original.isnull().sum().values,
    'Missing Percentage': (df_original.isnull().sum().values / len(df_original) * 100).round(2),
    'Data Type': df_original.dtypes.values
})

missing_analysis = missing_analysis.sort_values('Missing Count', ascending=False)
missing_analysis['Has Missing'] = missing_analysis['Missing Count'] > 0

print("\n📋 Missing Values Summary:")
display(missing_analysis)

total_missing = missing_analysis['Missing Count'].sum()
total_cells = df_original.shape[0] * df_original.shape[1]
print(f"\n📊 Overall Data Quality:")
print(f"   - Total cells: {total_cells:,}")
print(f"   - Total missing values: {total_missing:,}")
print(f"   - Overall completeness: {((total_cells - total_missing) / total_cells * 100):.2f}%")
```

```python
# Visualize missing values
missing_cols = missing_analysis[missing_analysis['Missing Count'] > 0]

if len(missing_cols) > 0:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Bar chart of missing counts
    colors = ['#e74c3c' if x > 5 else '#f39c12' if x > 2 else '#27ae60' 
              for x in missing_cols['Missing Percentage']]
    axes[0].barh(missing_cols['Column'], missing_cols['Missing Count'], color=colors)
    axes[0].set_xlabel('Number of Missing Values')
    axes[0].set_title('Missing Values by Column (Count)')
    for i, (count, pct) in enumerate(zip(missing_cols['Missing Count'], missing_cols['Missing Percentage'])):
        axes[0].text(count + 10, i, f'{count:,} ({pct}%)', va='center', fontsize=9)
    
    # Pie chart of data completeness
    complete = total_cells - total_missing
    axes[1].pie([complete, total_missing], 
                labels=['Complete', 'Missing'], 
                autopct='%1.2f%%',
                colors=['#27ae60', '#e74c3c'],
                explode=(0, 0.1))
    axes[1].set_title('Overall Data Completeness')
    
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in the dataset.")
```

### Data Cleaning Procedures

The following table documents all data cleaning steps performed:

```python
# Document cleaning procedures
cleaning_procedures = {
    'Variable': ['Item', 'Payment Method', 'Price', 'Quantity', 'Order Total', 'Order Date'],
    'Issue Found': [
        'Missing item names for some orders',
        'Missing payment method records',
        'Missing price values',
        'Missing quantity values',
        'Potential calculation errors after imputation',
        'String format, not datetime'
    ],
    'Handling Method': [
        'Filled with "Unknown" - preserves record for sales analysis',
        'Filled with "Unknown" - maintains payment trend analysis capability',
        'Imputed using median price within same Category',
        'Imputed using overall median quantity',
        'Recalculated as Price × Quantity',
        'Converted to datetime format using pd.to_datetime()'
    ],
    'Python Code': [
        "df['Item'].fillna('Unknown')",
        "df['Payment Method'].fillna('Unknown')",
        "df.groupby('Category')['Price'].transform(lambda x: x.fillna(x.median()))",
        "df['Quantity'].fillna(df['Quantity'].median())",
        "df['Order Total'] = df['Price'] * df['Quantity']",
        "pd.to_datetime(df['Order Date'])"
    ],
    'Justification': [
        'Unknown items can still contribute to category-level analysis',
        'Allows retention of sales data while flagging incomplete payment records',
        'Category median preserves relative pricing within food categories',
        'Median is robust to outliers; typical order quantity is stable',
        'Ensures Order Total accurately reflects imputed values',
        'Enables time-series analysis, day-of-week extraction'
    ]
}

df_cleaning = pd.DataFrame(cleaning_procedures)
print("=" * 100)
print("DATA CLEANING PROCEDURES DOCUMENTATION")
print("=" * 100)
display(df_cleaning)
```

```python
# Execute data cleaning with detailed comments
print("\n" + "=" * 60)
print("EXECUTING DATA CLEANING PROCEDURES")
print("=" * 60)

# Create a copy for cleaning
df_clean = df_original.copy()

# Step 1: Convert Order Date to datetime
print("\n[Step 1] Converting 'Order Date' to datetime format...")
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
print(f"   ✓ Date range: {df_clean['Order Date'].min().date()} to {df_clean['Order Date'].max().date()}")

# Step 2: Handle missing Item values
print("\n[Step 2] Handling missing 'Item' values...")
missing_items_before = df_clean['Item'].isnull().sum()
df_clean['Item'] = df_clean['Item'].fillna('Unknown')
print(f"   ✓ Filled {missing_items_before} missing items with 'Unknown'")

# Step 3: Handle missing Payment Method
print("\n[Step 3] Handling missing 'Payment Method' values...")
missing_payment_before = df_clean['Payment Method'].isnull().sum()
df_clean['Payment Method'] = df_clean['Payment Method'].fillna('Unknown')
print(f"   ✓ Filled {missing_payment_before} missing payment methods with 'Unknown'")

# Step 4: Handle missing Price values (using category median)
print("\n[Step 4] Handling missing 'Price' values using category median...")
missing_price_before = df_clean['Price'].isnull().sum()

# Calculate and display category medians for reference
category_medians = df_original.groupby('Category')['Price'].median()
print("   Category median prices used for imputation:")
for cat, median in category_medians.items():
    print(f"      - {cat}: ${median:.2f}")

df_clean['Price'] = df_clean.groupby('Category')['Price'].transform(
    lambda x: x.fillna(x.median())
)
print(f"   ✓ Filled {missing_price_before} missing prices with category medians")

# Step 5: Handle missing Quantity values
print("\n[Step 5] Handling missing 'Quantity' values...")
missing_qty_before = df_clean['Quantity'].isnull().sum()
overall_median_qty = df_original['Quantity'].median()
df_clean['Quantity'] = df_clean['Quantity'].fillna(overall_median_qty)
print(f"   ✓ Filled {missing_qty_before} missing quantities with median ({overall_median_qty})")

# Step 6: Recalculate Order Total
print("\n[Step 6] Recalculating 'Order Total'...")
df_clean['Order Total'] = df_clean['Price'] * df_clean['Quantity']
print(f"   ✓ Order Total recalculated for all {len(df_clean)} records")

# Step 7: Add derived time features
print("\n[Step 7] Adding derived time features...")
df_clean['Year'] = df_clean['Order Date'].dt.year
df_clean['Month'] = df_clean['Order Date'].dt.month
df_clean['Day of Week'] = df_clean['Order Date'].dt.day_name()
df_clean['Quarter'] = df_clean['Order Date'].dt.quarter
print("   ✓ Added: Year, Month, Day of Week, Quarter")

print("\n" + "=" * 60)
print("DATA CLEANING COMPLETE!")
print("=" * 60)
```

```python
# Verify cleaning results
print("\n" + "=" * 60)
print("POST-CLEANING VERIFICATION")
print("=" * 60)

print("\n📋 Missing Values After Cleaning:")
print(df_clean.isnull().sum())

print("\n📊 Cleaned Dataset Info:")
print(f"   - Total Records: {len(df_clean):,}")
print(f"   - Total Columns: {len(df_clean.columns)}")
print(f"   - New Columns Added: Year, Month, Day of Week, Quarter")

print("\n📈 Cleaned Dataset Sample:")
display(df_clean.head(10))
```

```python
# Create mapping dictionaries for merging with main dataset
taste_profile_map = {row['Item']: row['Taste_Profile'] for _, row in menu_df.iterrows()}
temperature_map = {row['Item']: row['Temperature'] for _, row in menu_df.iterrows()}
health_level_map = {row['Item']: row['Health_Level'] for _, row in menu_df.iterrows()}
ingredient_type_map = {row['Item']: row['Ingredient_Type'] for _, row in menu_df.iterrows()}

# Add classification columns to the cleaned dataset
print("\n" + "=" * 60)
print("ADDING MENU CLASSIFICATION TO DATASET")
print("=" * 60)

df_clean['Taste_Profile'] = df_clean['Item'].map(taste_profile_map)
df_clean['Temperature'] = df_clean['Item'].map(temperature_map)
df_clean['Health_Level'] = df_clean['Item'].map(health_level_map)
df_clean['Ingredient_Type'] = df_clean['Item'].map(ingredient_type_map)

print("\n✓ Added columns: Taste_Profile, Temperature, Health_Level, Ingredient_Type")
print(f"\nUpdated dataset shape: {df_clean.shape}")
print(f"\nNew columns: {list(df_clean.columns[-4:])}")
```

```python
# Visualize menu classification distribution
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Health Level Distribution
health_counts = df_clean['Health_Level'].value_counts()
colors_health = {'Healthy': '#27ae60', 'Moderate': '#f39c12', 'Indulgent': '#e74c3c', 'Unknown': '#95a5a6'}
axes[0, 0].pie(health_counts.values, labels=health_counts.index, autopct='%1.1f%%',
               colors=[colors_health.get(x, '#95a5a6') for x in health_counts.index])
axes[0, 0].set_title('Orders by Health Level', fontsize=12, fontweight='bold')

# 2. Temperature Distribution
temp_counts = df_clean['Temperature'].value_counts()
colors_temp = {'Hot': '#e74c3c', 'Cold': '#3498db', 'Room Temp': '#f39c12', 'Unknown': '#95a5a6'}
axes[0, 1].pie(temp_counts.values, labels=temp_counts.index, autopct='%1.1f%%',
               colors=[colors_temp.get(x, '#95a5a6') for x in temp_counts.index])
axes[0, 1].set_title('Orders by Temperature Preference', fontsize=12, fontweight='bold')

# 3. Taste Profile Distribution
taste_counts = df_clean['Taste_Profile'].value_counts().head(8)
taste_counts.plot(kind='barh', ax=axes[1, 0], color='steelblue')
axes[1, 0].set_xlabel('Number of Orders')
axes[1, 0].set_title('Orders by Taste Profile (Top 8)', fontsize=12, fontweight='bold')

# 4. Ingredient Type Distribution
ingredient_counts = df_clean['Ingredient_Type'].value_counts().head(8)
ingredient_counts.plot(kind='barh', ax=axes[1, 1], color='coral')
axes[1, 1].set_xlabel('Number of Orders')
axes[1, 1].set_title('Orders by Ingredient Type (Top 8)', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Health Level vs Category Analysis
print("\n" + "=" * 60)
print("HEALTH LEVEL BY CATEGORY ANALYSIS")
print("=" * 60)

health_category = pd.crosstab(df_clean['Category'], df_clean['Health_Level'], normalize='index') * 100
health_category = health_category.round(1)

print("\nPercentage of orders by Health Level within each Category:")
display(health_category)

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(health_category, annot=True, fmt='.1f', cmap='RdYlGn_r', cbar_kws={'label': 'Percentage (%)'})
plt.title('Health Level Distribution by Category (%)', fontsize=12, fontweight='bold')
plt.xlabel('Health Level')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

```python
# Save the final enriched dataset
print("\n" + "=" * 60)
print("SAVING ENRICHED DATASET")
print("=" * 60)

# Define final column order
final_columns = [
    'Order ID', 'Customer ID', 'Order Date', 'Year', 'Month', 'Quarter', 'Day of Week',
    'Category', 'Item', 'Taste_Profile', 'Temperature', 'Health_Level', 'Ingredient_Type',
    'Price', 'Quantity', 'Order Total', 'Payment Method'
]

df_final = df_clean[final_columns]

# Save to CSV
df_final.to_csv('restaurant_sales_data_enriched.csv', index=False)
print("\n✓ Saved: restaurant_sales_data_enriched.csv")

# Save menu classification reference table
menu_df.to_csv('menu_classification_reference.csv', index=False)
print("✓ Saved: menu_classification_reference.csv")

# Also save to Excel with multiple sheets
with pd.ExcelWriter('the_bistro_data.xlsx', engine='openpyxl') as writer:
    df_final.to_excel(writer, sheet_name='Sales Data', index=False)
    menu_df.to_excel(writer, sheet_name='Menu Classification', index=False)
    df_location.to_excel(writer, sheet_name='Location Profile', index=False)
    df_customer.to_excel(writer, sheet_name='Customer Profile', index=False)
    df_cleaning.to_excel(writer, sheet_name='Data Cleaning Log', index=False)

print("✓ Saved: the_bistro_data.xlsx (with multiple sheets)")

print(f"\n📊 Final Dataset Summary:")
print(f"   - Total Records: {len(df_final):,}")
print(f"   - Total Columns: {len(df_final.columns)}")
print(f"   - Date Range: {df_final['Order Date'].min()} to {df_final['Order Date'].max()}")
```
